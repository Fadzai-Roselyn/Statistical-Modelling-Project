# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import pandas as pd 
import requests
import json
```

```python
# Exploring City Bikes API trying to chose city to get bike info from

# defining Networks URl for citybike
url = f"http://api.citybik.es/v2/networks"

#requesting API to give all networks data for CityBikes
response = requests.get(url)

if response.status_code == 200:
    print("Request Made Successfully")
    
else:
    print("Error Requesting Data:", response.status_code)
```

```python
# Parsing data to Json format
networks_data = response.json()
```

```python
# Choosing a city
city_name = "Vancouver"  # defining city_name

# creating url to select Vancouver as city from networks 
url_city = f"http://api.citybik.es/v2/networks/mobibikes"
response_city = requests.get(url_city)

if response_city.status_code == 200:
       # parsing the request data into json if the request is successfull
         Vancouver_city = response_city.json()
         print("Successful request and parsing")
else:
         print("Request not Successful:", response_city.status_code)
```

```python
# Exploring the data
Vancouver_city
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
all_stations = Vancouver_city['network'].get('stations')
```

```python
if all_stations:
    stations = []
    for station in all_stations:
        station_name = station.get('name')
        latitude    =  station.get('latitude')
        longitude   =  station.get('longitude')
        bikes_available = station.get('free_bikes')
        stations.append({
             'Station Name': station_name,
             'Latitude': latitude,
             'Longitude': longitude,
             'Bikes Available': bikes_available
              })
```

Put your parsed results into a DataFrame.

```python
Bike_stations = pd.DataFrame(stations)
Bike_stations
```

```python
Bike_stations.to_csv('bikes_df.csv', index=False)
```
